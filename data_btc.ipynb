```python
import requests
import pandas as pd
import datetime

# Define a data de início e a data de hoje
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime.now()

# Lista para armazenar os resultados
results = []

# Configuração da API
api_url = 'https://api.binance.com/api/v3/klines'  # Substitua pela URL da sua API
symbol = 'BTCBRL'
interval = '1m'
limit = 1440

# Itera sobre cada dia entre start_date e end_date
current_date = start_date
while current_date <= end_date:
    # Converte as datas em milissegundos
    start_time = int(current_date.timestamp() * 1000)
    end_time = int((current_date + datetime.timedelta(days=1)).timestamp() * 1000)

    # Faz a requisição à API
    params = {'symbol': symbol, 'interval': interval, 'startTime': start_time, 'endTime': end_time, 'limit': limit}
    response = requests.get(api_url, params=params)
    
    if response.status_code == 200:
        # Supondo que a resposta da API é uma lista de dicionários
        api_data = response.json()  # Ajuste conforme necessário
        
        # Adiciona os dados ao resultado
        results.extend(api_data)  # Adiciona todos os dados da resposta
    else:
        print(f"Erro ao acessar a API para a data {current_date.strftime('%Y-%m-%d')}: {response.status_code}")
        
    time.sleep(1)
    # Avança para o próximo dia
    current_date += datetime.timedelta(days=1)

# Cria um DataFrame com os resultados
df = pd.DataFrame(results)
# Salva os resultados em um arquivo CSV
df.to_csv('brlbtc.csv', index=False)
print("Dados salvos em resultados.csv")
```

```python
import os
import shutil
import zipfile

def organizar_arquivos_zip(pasta_origem, pasta_destino_zip, pasta_destino_extracao):
    """
    Mapeia, move e extrai arquivos ZIP.

    Args:
        pasta_origem: Caminho da pasta onde os arquivos ZIP estão localizados.
        pasta_destino_zip: Caminho da pasta para mover os arquivos ZIP.
        pasta_destino_extracao: Caminho da pasta para extrair o conteúdo dos ZIPs.
    """

    for raiz, diretorios, arquivos in os.walk(pasta_origem):
        for arquivo in arquivos:
            if arquivo.endswith(".zip"):
                # Caminho completo do arquivo
                print(arquivo)
                caminho_completo = os.path.join(raiz, arquivo)

                # Mover o arquivo para a pasta de destino
                novo_caminho = os.path.join(pasta_destino_zip, arquivo)
                shutil.move(caminho_completo, novo_caminho)

                # Extrair o arquivo para a pasta de extração
                with zipfile.ZipFile(novo_caminho, 'r') as zip_ref:
                    zip_ref.extractall(pasta_destino_extracao)

# Exemplo de uso
pasta_origem = "C:/Users/Mariano/Projetos/deepQLearning/BTC"  # Substitua pelo seu caminho
pasta_destino_zip = "C:/Users/Mariano/Projetos/deepQLearning/BTC"
pasta_destino_extracao = "C:/Users/Mariano/Projetos/deepQLearning/EXTRACT_BTC"

organizar_arquivos_zip(pasta_origem, pasta_destino_zip, pasta_destino_extracao)
```

```python
import pandas as pd
import datetime
import os

def excel_para_csv(diretorio, nome_arquivo_saida):
    """
    Converte todos os arquivos Excel em um diretório para um único arquivo CSV.

    Args:
        diretorio: Caminho para o diretório com os arquivos Excel.
        nome_arquivo_saida: Nome do arquivo CSV de saída.
    """

    df_final = pd.DataFrame()

    for arquivo in os.listdir(diretorio):
        if arquivo.endswith('.xlsx'):
            arquivo
            caminho_completo = os.path.join(diretorio, arquivo)
            df_temp = pd.read_excel(caminho_completo)
            df_temp['Date'] = datetime.datetime.strptime(arquivo[-13:-5], "%Y%m%d").strftime("%Y-%m-%d")
            df_final = pd.concat([df_final, df_temp], ignore_index=True)
            
    df_final.to_csv(nome_arquivo_saida, index=False)

# Exemplo de uso
diretorio_excel = "C:/Users/Mariano/Projetos/deepQLearning/EXTRACT_BTC"  # Substitua pelo caminho do seu diretório
nome_arquivo_saida = "brlbtc.csv"

excel_para_csv(diretorio_excel, nome_arquivo_saida)
```
